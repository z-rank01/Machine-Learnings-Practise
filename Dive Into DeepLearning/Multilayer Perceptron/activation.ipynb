```python
# activation function visualization
import torch
import matplotlib.pyplot as plt

figsize = (6, 3)

# Rectified Linear Units, ReLU
# -------------------------------
X = torch.arange(-8.0, 8.0, 0.1, requires_grad=True)
y = torch.relu(X)
fig, axes = plt.subplots(1, 2, figsize=figsize)
axes[0].plot(X.detach(), y.detach())
axes[0].set_title('ReLU - function')
axes[0].set_xlabel('x')
axes[0].set_ylabel('ReLU(x)')
y.backward(torch.ones_like(X), retain_graph=True)
axes[1].plot(X.detach(), X.grad)
axes[1].set_title('ReLU - Gradient')
axes[1].set_xlabel('x')
axes[1].set_ylabel('Grad(ReLU(x))')
plt.tight_layout()
plt.show()
```

```python
# Parameterized Rectified Linear Units, pReLU
# -------------------------------
X = torch.arange(-8.0, 8.0, 0.1, requires_grad=True)
y = torch.prelu(X, torch.tensor(0.2))
fig, axes = plt.subplots(1, 2, figsize=figsize)
axes[0].plot(X.detach(), y.detach())
axes[0].set_title('pReLU - function')
axes[0].set_xlabel('x')
axes[0].set_ylabel('pReLU(x)')
y.backward(torch.ones_like(X), retain_graph=True)
axes[1].plot(X.detach(), X.grad)
axes[1].set_title('pReLU - Gradient')
axes[1].set_xlabel('x')
axes[1].set_ylabel('Grad(pReLU(x))')
plt.tight_layout()
plt.show()
```

```python
# sigmoid
# -------
X = torch.arange(-8.0, 8.0, 0.1, requires_grad=True)
y = torch.sigmoid(X)
fig, axes = plt.subplots(1, 2, figsize=figsize)
axes[0].plot(X.detach(), y.detach())
axes[0].set_title('sigmoid - function')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sigmoid(x)')
y.backward(torch.ones_like(X), retain_graph=True)
axes[1].plot(X.detach(), X.grad)
axes[1].set_title('sigmoid - Gradient')
axes[1].set_xlabel('x')
axes[1].set_ylabel('Grad(sigmoid(x))')
plt.tight_layout()
plt.show()
```
