```python
# Density-Based Spatial Clustering of Applications with Noise, DBSCAN

# compare k-means, agglomerative clustering and DBSCAN 
# on non-spherical datasets
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
import matplotlib.pyplot as plt

# set up dataset
X, y = make_moons(n_samples=500, noise=0.1, random_state=0)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# train model
km = KMeans(n_clusters=2, random_state=0)
ac = AgglomerativeClustering(n_clusters=2, 
                             affinity='euclidean', 
                             linkage='complete')
db = DBSCAN(eps=0.2, 
            min_samples=10,   # adjust/tune relied on samples scale
            metric='euclidean')
y = [km.fit_predict(X), ac.fit_predict(X), db.fit_predict(X)]
cluster_name = ['K-means', 'Agglomerative clustering', 'DBSCAN']
fig, ax = plt.subplots(1, 3, figsize=(16, 4))

for i in range(0, 3):
    ax[i].scatter(X[y[i]==0, 0], 
                  X[y[i]==0, 1], 
                  c='lightblue', 
                  edgecolors='black', 
                  marker='o', 
                  s=40, 
                  label='Cluster 1')
    ax[i].scatter(X[y[i]==1, 0], 
                  X[y[i]==1, 1], 
                  c='lightgreen', 
                  edgecolors='black', 
                  marker='s', 
                  s=40, 
                  label='Cluster 2')
    ax[i].set_title(cluster_name[i])
plt.legend(loc='upper right')
plt.show()
```
