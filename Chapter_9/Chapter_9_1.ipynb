```python
# K-means

# set up a datasets
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=150, 
                  n_features=2, 
                  centers=3, 
                  cluster_std=0.5, 
                  shuffle=True,
                  random_state=0)

# visualize cluster datasets
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], 
            X[:, 1], 
            c='white', 
            marker='o', 
            edgecolors='black', 
            s=50)
plt.grid()
plt.show()
```

```python
# k-means algorithm
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, 
            n_init=10, 
            init='random', 
            max_iter=10, 
            tol=0.0001, 
            random_state=0)
y_pred = km.fit_predict(X)

plt.scatter(X[y_pred==0, 0], 
            X[y_pred==0, 1], 
            s=50, c='lightgreen', 
            marker='o', edgecolors='black', 
            label='Cluster 1')
plt.scatter(X[y_pred==1, 0], 
            X[y_pred==1, 1], 
            s=50, c='orange', 
            marker='o', edgecolors='black', 
            label='Cluster 2')
plt.scatter(X[y_pred==2, 0], 
            X[y_pred==2, 1], 
            s=50, c='lightblue', 
            marker='o', edgecolors='black', 
            label='Cluster 3')
plt.scatter(km.cluster_centers_[:, 0], 
            km.cluster_centers_[:, 1], 
            s=250, marker='*', 
            c='red', edgecolors='black', 
            label='Centroids')
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```python
# k-means++ algorithm
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, 
            n_init=10, 
            init='k-means++',    # for better k-means++ algorithm, we initialize with 'k-means++' to activate it.
            max_iter=10, 
            tol=0.0001, 
            random_state=0)
y_pred = km.fit_predict(X)

plt.scatter(X[y_pred==0, 0], 
            X[y_pred==0, 1], 
            s=50, c='lightgreen', 
            marker='o', edgecolors='black', 
            label='Cluster 1')
plt.scatter(X[y_pred==1, 0], 
            X[y_pred==1, 1], 
            s=50, c='orange', 
            marker='o', edgecolors='black', 
            label='Cluster 2')
plt.scatter(X[y_pred==2, 0], 
            X[y_pred==2, 1], 
            s=50, c='lightblue', 
            marker='o', edgecolors='black', 
            label='Cluster 3')
plt.scatter(km.cluster_centers_[:, 0], 
            km.cluster_centers_[:, 1], 
            s=250, marker='*', 
            c='red', edgecolors='black', 
            label='Centroids')
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```
